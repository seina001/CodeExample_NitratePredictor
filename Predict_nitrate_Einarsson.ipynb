```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#import the unkwown wine file, as well as the wine dataset from sklearn
infile = 'jma_nitrate.csv'
nitratefile = pd.read_csv(infile,sep=(','))
nitratefile = nitratefile[nitratefile.nitrate>0]
X = nitratefile[["sal","theta"]]
y = nitratefile[["nitrate"]]
print(np.shape(X))
print(np.shape(y))



from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
# generate the data
import matplotlib.pyplot as plt
import random

n  = int(input('Choose a number for k: '))

# generate some mostly linear data with stochastic noise added


# pick model
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=n)
model.fit(X_train, y_train)

# compute the R^2 score
print("Training R^2 score: {0:.2f}".format(model.score(X_train,y_train)))

# plot the model together with the data
xfit = X_test
yfit = model.predict(xfit)
print("Testing R^2 score: {0:.2f}".format(model.score(xfit,yfit)))

#plt.scatter(X_train[["sal"]],y_train)
plt.scatter(X_train[["theta"]],y_train)
plt.scatter(xfit[["theta"]], yfit,c = 'k')
#plt.show()
#from scipy.spatial import distance
#euc = distance.euclidean(xfit, yfit)
#print euc
```
